```python
sales['Item_Weight'].fillna(0, inplace= True)
```

```python
item_avg_vsb = sales.pivot_table(values='Item_Visibility', index= 'Item_Type')
item_avg_vsb
```

```python
for i in range(0,8522):
    if (sales['Item_Visibility'][i]==0):
        w = item_avg_vsb[item_avg_vsb.index == sales['Item_Type'][i]]['Item_Visibility']
        sales['Item_Visibility'][i] = w
```

```python
sales["Item_Fat_Content"].unique()
```

```python
for i in range(0,8522):
    if sales['Item_Fat_Content'][i] == 'LF':
        sales['Item_Fat_Content'][i] = 'Low Fat'
    elif sales['Item_Fat_Content'][i] == 'low fat':
        sales['Item_Fat_Content'][i] = 'Low Fat'
    elif sales['Item_Fat_Content'][i] == 'reg':
        sales['Item_Fat_Content'][i] = 'Regular'
            
```

```python
sales['Item_Fat_Content'].unique()
```

```python
sales['Outlet_Size'].fillna(0, inplace= True)
```

```python
for i in range(0,8522):
    if sales['Outlet_Size'][i]==0:
        if sales['Outlet_Type'][i]=="Grocery Store":
            sales['Outlet_Size'][i]='Small'
        else:
            sales['Outlet_Size'][i]='Medium'
```

```python
sales['Outlet_Size'].unique()
```

```python
sales['Item_Type'].unique()
```

```python
for i in range(0,8522):
    if sales['Item_Type'][i] in ['Fruits and Vegetables','Snack Foods', 'Meat', 'Baking Goods', 'Frozen Foods', 'Starchy Foods', 'Canned', 'Dairy', 'Breads', 'Breakfast']:
        sales['Item_Type'][i]='Food'
    elif sales['Item_Type'][i] in ['Household', 'Health and Hygiene']:
        sales['Item_Type'][i]='Non-consumables'
    elif sales['Item_Type'][i] in ['Soft Drinks', 'Hard Drinks']:
        sales['Item_Type'][i]='Drinks'
```

```python
sales['Item_Type'][8522]='Drinks'
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

sales['Outlet'] = le.fit_transform(sales['Outlet_Identifier'])
var_mod = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type','Outlet_Type','Outlet']
le = LabelEncoder()
for i in var_mod:
    sales[i] = le.fit_transform(sales[i])
```

```python
sales['Item_Type'].unique()
```

```python
sales = pd.get_dummies(sales, columns=['Outlet_Location_Type','Outlet_Size','Outlet_Type',
                              'Item_Type','Outlet'])
```

```python
sales.dtypes
```

```python
sales['Age']=0
for i in range(0,8523):
    sales['Age'][i]= 2019- (sales['Outlet_Establishment_Year'][i])
```

```python
sales.drop(['Item_Identifier'], axis=1, inplace=True)
```

```python
sales.drop(['Outlet_Identifier'], axis=1 , inplace=True)
```
